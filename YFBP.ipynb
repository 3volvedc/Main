```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime
from statsmodels.formula.api import ols
import math
```

```python
def cat_codes(df, column, new_column):
    for column in df.columns:
        if df[column].dtype == 'category':
            df[new_column] = df[column].cat.codes
    return df


def int_values(df, column_name):
    df[column_name] = abs(df[column_name].round().astype(int))
    return df


def to_datetime(dataset, column):
    new_datetimes = []
    dataset[column] = dataset[column].astype('string')
    dataset[column] = pd.to_datetime(dataset[column])
    return dataset


def set_colors_sizes(dataset, column):
    first = np.quantile(dataset[column], .25).round().astype(int)
    second = np.quantile(dataset[column], .50).round().astype(int)
    third = np.quantile(dataset[column], .75).round().astype(int)
    max = dataset[column].max()
    
    colors, quantile = [], []
    
    for lab, row in shows.iterrows():
        if row[column] <= first:
            colors.append("red")
            quantile.append("1st")
        elif third > row[column] <= second:
            colors.append("blue")
            quantile.append("2nd")
        elif max > row[column] <= third:
            colors.append("orange")
            quantile.append("3rd")
        elif row[column] == max:
            colors.append("gold")
            quantile.append("max")
        else:
            colors.append("green")
            quantile.append("4th")
    
    dataset['visual_color'] = colors
    dataset['views_quantile'] = quantile
    display([first, second, third, max])
    return dataset


def concat_3(data1, data2, data3):
    new_concat_df = pd.concat([data1, data2, data3], axis=1)
    new_concat_df.rename(columns={0: 'Views per Minute (vpm)'}, inplace=True)
    return new_concat_df

def scatter_2_regplot(x, y, dataset1, dataset2, dataset3, label1, label2, label3):
    fig = plt.figure()
    plt.figure(figsize=(15, 6))
    sns.set_style('ticks')
    sns.scatterplot(x=x, y=y, data=dataset1, color='gold', label=label1)
    sns.scatterplot(x=x, y=y, data=dataset2, color='red', label=label2)
    sns.regplot(x=x, y=y, data=dataset3, color='green', label=label3)
    plt.legend()
    return plt.show()
```

```python
avg_views_weekday = shows.groupby(['weekday'])['Views'].mean().round().astype('int')

avg_views_year = shows.groupby(['year'])['Views'].mean().round().astype('int')

avg_views_month = shows.groupby(['month'])['Views'].mean().round().astype('int')
avg_length_month = shows.groupby(['month'])['Length'].mean().round().astype('int')
avg_vpm_month = (avg_views_month / avg_length_month).round().astype('int')

#concat_3(data1, data2, data3)
concat_3(avg_views_month, avg_length_month, avg_vpm_month)
```

```python
samples = shows.sample(frac=1/10, replace=False)
samples
```

```python
mdl_views_v_guests = ols('Views ~ vpm', data=shows).fit()
mdl_views_v_guests.params
```

```python
explanatory_data = pd.DataFrame({'vpm': np.arange(shows['vpm'].min(), shows['vpm'].max(), 11)})
len(explanatory_data)
```

```python
prediction_data = explanatory_data.assign(Views = mdl_views_v_guests.predict(explanatory_data))

#int_values(df, column_name)
prediction_data = int_values(prediction_data, 'Views')
prediction_data.tail(10)
```
